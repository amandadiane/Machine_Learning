#### Splitting the data: training and test 

```python
from sklearn.model_selection import train_test_split

X = bc.drop(["id", "diagnosis"], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

## 1. Support Vector Machine (SVM) (20 pts)

**Question (15pts):** Train and test a SVM model with the following requirements:
- Use GridSearchCV to find the best combination of model hyper-parameters: kernel: {linear, rbf}; for both kernel, C:{1, 10, 100, 1000}; for rbf kernel 'gamma' = {1e-3, 1e-4, 1e-5}.
- Try both accuracy and recall metric as scoring function in GridSearchCV
- Report the best hyper-parameter setting for each metric
- Report and plot the confusion matrix of the best estimator for each metric

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from itertools import product
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix

svc= SVC()
param_grid_linear = {"C":[1, 10, 100, 1000], "kernel":["linear"]}
param_grid_rbf = {"C":[1, 10, 100, 1000], "gamma":[1e-3, 1e-4, 1e-5], "kernel":["rbf"]}
metrics = ["accuracy", "recall"]
kernels = ["linear", "rbf"]
matrixes = []    
    
for sc in metrics:
    temp_scores ={}
    for ker in kernels:
        if ker == "linear":
            grid= GridSearchCV(svc, param_grid=param_grid_linear, scoring=sc)
            print("# Tuning hyper-parameters for {} and {}".format(sc, ker))
            grid.fit(X_train,y_train)
            svc_best_param = grid.best_params_
            svc_best_score = grid.best_score_
            temp_scores["linear"] = [svc_best_param, svc_best_score]
            print("best params for ", sc, ker, svc_best_param)
        else:
            grid= GridSearchCV(svc, param_grid=param_grid_rbf, scoring=sc)
            print("# Tuning hyper-parameters for {} and {}".format(sc, ker))
            grid.fit(X_train,y_train)
            svc_best_param = grid.best_params_
            temp_scores["rbf"] = svc_best_param
            svc_best_score = grid.best_score_
            temp_scores["rbf"] = [svc_best_param, svc_best_score]
            print("best params for ", sc, ker, svc_best_param)
    print("\n", temp_scores.get("linear"), temp_scores.get("rbf"))
    best_ = np.argmax([temp_scores.get("linear")[1],temp_scores.get("rbf")[1]])
    best_model = [temp_scores["linear"] if best_ == 0 else temp_scores["rbf"]]
    params = dict(best_model[0][0])
    print("The best model is {} with a score {} \n\n\n".format(best_model[0][0], best_model[0][1]))
    if params.get("gamma")== None:
        model_final = SVC(kernel=params["kernel"], C=params["C"])
    else:
        model_final = SVC(kernel=params["kernel"], C=params["C"], gamma=params.get("gamma"))
    model_final.fit(X_train, y_train)
    y_pred_svc = model_final.predict(X_test)
    matrix = confusion_matrix(y_test, y_pred_svc)
    matrixes.append(matrix)

plt.subplot(2, 1, 1)    
sns.heatmap(matrixes[0].T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label')
plt.subplot(2, 1, 2)
sns.heatmap(matrixes[1].T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label')
plt.show()
```

```python
from sklearn.neighbors.kde import KernelDensity
D = [[0],[1],[1],[2],[3],[4],[4],[4],[5]]
kde = KernelDensity(bandwidth = 0.1).fit(D)
D_den = kde.score_samples(D)
```

```python
kde2 = KernelDensity(bandwidth = 1).fit(D)
D_den2 = kde2.score_samples(D)
```
